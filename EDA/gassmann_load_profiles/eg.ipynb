```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('/Users/wastechs/Documents/git-repos/energy-efficiency')
from lib.util.helper import weekday_time_series
import seaborn as sns
import plotly.express as px
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf
```

```python
eg = weekday_time_series(sensor_id='eg_10T')
```

```python
px.line(
    x=eg.index, y=eg.kw, 
    title='EG',
    labels={
        'x': 'Time',
        'y': 'kW'
    })
```

```python
time_int_range = np.arange(0, eg.shape[0]*10, 10)
eg['t'] = time_int_range
eg['t_scaled'] = (eg['t'] - eg['t'].min()) / (eg['t'].max() - eg['t'].min())
```

```python
px.line(
    x=eg['t_scaled'], y=eg.kw, 
    title='EG',
    labels={
        'x': 'Time',
        'y': 'kW'
    })
```

```python
fig, ax = plt.subplots(figsize=(16, 7))
plot_acf(eg.kw.values.squeeze(), lags=160, ax=ax)
```
