```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('/Users/wastechs/Documents/git-repos/energy-efficiency')
from lib.util.helper import query_table
import seaborn as sns
import plotly.express as px
from statsmodels.graphics.tsaplots import plot_acf
```

```python
df = query_table(table='uv_sigma_line_eg_15T')
```

```python
px.line(
    x=df.index, y=df.kw, 
    title='UV Sigma Line',
    labels={
        'x': 'Time',
        'y': 'kW'
    })
```

```python
import numpy as np

time_int_range = np.arange(0, df.shape[0]*15, 15)

df['t'] = time_int_range

df['t_scaled'] = (df['t'] - df['t'].min()) / (df['t'].max() - df['t'].min())

px.line(
    x=df['t_scaled'], y=df.kw, 
    title='UV EG',
    labels={
        'x': 'Time',
        'y': 'kW'
    })
```

```python
fig, ax = plt.subplots(figsize=(16, 7))
plot_acf(df.kw.values.squeeze(), lags=100, ax=ax)
```
